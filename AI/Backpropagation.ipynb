```python
import math

# 시그모이드 함수 정의
def sigmoid(v):
    return 1 / (1 + math.exp(-v))

# 시그모이드 도함수 정의
def desigmoid(v):
    return v * (1 - v)

# 가중치 정의
w11 = 0.74
w12 = 0.75
w13 = 0.61

w21 = 0.22
w22 = 0.5
w23 = 0.23

w14 = 0.24
w15 = 0.14

w24 = 0.29
w25 = 0.29

w34 = 0.5
w35 = 0.9

w46 = 0.43
w56 = 0.39

# 매개변수 출력
print("\n초기 매개변수")
print("w11: ", w11)
print("w12: ", w12)
print("w13: ", w13)

print("w21: ", w21)
print("w22: ", w22)
print("w23: ", w23)

print("w14: ", w14)
print("w15: ", w15)

print("w24: ", w24)
print("w25: ", w25)

print("w34: ", w34)
print("w35: ", w35)

print("w46: ", w46)
print("w56: ", w56)

# 초기 값 정의
x1 = 1
x2 = 0
t = 1 # result
eta = 0.1

# 활성 값 계산
y1 = sigmoid(w11 * x1 + w21 * x2)
y2 = sigmoid(w12 * x1 + w22 * x2)
y3 = sigmoid(w13 * x1 + w23 * x2)
y4 = sigmoid(w14 * y1 + w24 * y2 + w34 * y3)
y5 = sigmoid(w15 * y1 + w25 * y2 + w35 * y3)
y = sigmoid(w46 * y4 + w56 * y5)

print("\n초기 결과값:", y)

# 오차값 계산
o = (t - y) * desigmoid(y)
o5 = desigmoid(y5) * w56 * o
o4 = desigmoid(y4) * w46 * o
o3 = desigmoid(y3) * (w34 * o4 + w35 * o5)
o2 = desigmoid(y2) * (w24 * o4 + w25 * o5)
o1 = desigmoid(y1) * (w14 * o4 + w15 * o5)

# 오차 값 출력
print("\n오차값")
print("o: ", o)
print("o5: ", o5)
print("o4: ", o4)
print("o3: ", o3)
print("o2: ", o2)
print("o1: ", o1)

# 가중치 수정
w11 += eta * x1 * o1
w12 += eta * x1 * o2
w13 += eta * x1 * o3

w21 += eta * x2 * o1
w22 += eta * x2 * o2
w23 += eta * x2 * o3

w14 += eta * y1 * o4
w15 += eta * y1 * o5

w24 += eta * y2 * o4
w25 += eta * y2 * o5

w34 += eta * y3 * o4
w35 += eta * y3 * o5

w46 += eta * y4 * o
w56 += eta * y5 * o

# 매개변수 출력
print("\n변경된 매개변수")
print("w11: ", w11)
print("w12: ", w12)
print("w13: ", w13)

print("w21: ", w21)
print("w22: ", w22)
print("w23: ", w23)

print("w14: ", w14)
print("w15: ", w15)

print("w24: ", w24)
print("w25: ", w25)

print("w34: ", w34)
print("w35: ", w35)

print("w46: ", w46)
print("w56: ", w56)

# 활성 값 재계산
y1 = sigmoid(w11 * x1 + w21 * x2)
y2 = sigmoid(w12 * x1 + w22 * x2)
y3 = sigmoid(w13 * x1 + w23 * x2)
y4 = sigmoid(w14 * y1 + w24 * y2 + w34 * y3)
y5 = sigmoid(w15 * y1 + w25 * y2 + w35 * y3)
y = sigmoid(w46 * y4 + w56 * y5)

print("\n변경된 결과값:", y)
```
