```python
import os
fname = os.path.join("jena_climate_2009_2016.csv")

with open(fname) as f:
    data = f.read()

lines = data.split("\n")
header = lines[0].split(",")
lines = lines[1:]
print(header)
print(len(lines))
```

```python
import numpy as np

temperature = np.zeros((len(lines),)) ## 온도를 나타내는 numpy 배열 생성
raw_data = np.zeros((len(lines), len(header) - 1)) ## numpy 배열 생성
for i, line in enumerate(lines):
    values = [float(x) for x in line.split(",")[1:]]
    temperature[i] = values[1] ## 두 번째 열을 'temperature'배열에 저장
    raw_data[i, :] = values[:] ## 모든 열을 'raw_data'배열에 저장
```

**전체 기온을 그래프로 그리기**

```python
from matplotlib import pyplot as plt
plt.plot(range(len(temperature)), temperature)
plt.show()
```

**처음 10일간의 기온을 그래프로 그리기**

```python
plt.plot(range(1440), temperature[:1440])
plt.show()
```

**각 분할에 사용할 샘플 수 계산하기**

```python
num_train_samples = int(0.5 * len(raw_data))
num_val_samples = int(0.25 * len(raw_data))
num_test_samples = len(raw_data) - num_train_samples - num_val_samples
print("num_train_samples:", num_train_samples)
print("num_val_samples:", num_val_samples)
print("num_test_samples:", num_test_samples)
```

### 데이터 준비

**데이터 정규화**

```python
mean = raw_data[:num_train_samples].mean(axis=0)
raw_data -= mean
std = raw_data[:num_train_samples].std(axis=0)
raw_data /= std
```

```python
import numpy as np
from tensorflow import keras
int_sequence = np.arange(10)
dummy_dataset = keras.utils.timeseries_dataset_from_array(
    data=int_sequence[:-3],
    targets=int_sequence[3:],
    sequence_length=3, ## 시퀀스의 길이는 3스텝
    batch_size=2, ## 시퀀스의 배치 크기는 2
)

for inputs, targets in dummy_dataset: ## 6개의 생성된 시퀀스
    for i in range(inputs.shape[0]): ## inputs.shape[0] : (2, 3)(batch_size=2), i=0,1
        print([int(x) for x in inputs[i]], int(targets[i]))
```

**훈련, 검증, 테스트 데이터셋 만들기**

```python
sampling_rate = 6
sequence_length = 120
delay = sampling_rate * (sequence_length + 24 - 1)
batch_size = 256

train_dataset = keras.utils.timeseries_dataset_from_array(
    raw_data[:-delay],
    targets=temperature[delay:],
    sampling_rate=sampling_rate,
    sequence_length=sequence_length,
    shuffle=True,
    batch_size=batch_size,
    start_index=0,
    end_index=num_train_samples)

val_dataset = keras.utils.timeseries_dataset_from_array(
    raw_data[:-delay],
    targets=temperature[delay:],
    sampling_rate=sampling_rate,
    sequence_length=sequence_length,
    shuffle=True,
    batch_size=batch_size,
    start_index=num_train_samples,
    end_index=num_train_samples + num_val_samples)

test_dataset = keras.utils.timeseries_dataset_from_array(
    raw_data[:-delay],
    targets=temperature[delay:],
    sampling_rate=sampling_rate,
    sequence_length=sequence_length,
    shuffle=True,
    batch_size=batch_size,
    start_index=num_train_samples + num_val_samples)
```

**훈련 데이터셋의 배치 크기 확인하기**

```python
for samples, targets in train_dataset:
    print("샘플 크기:", samples.shape)
    print("타깃 크기:", targets.shape)
    break
```

### 상식 수준의 기준점

**상식적인 기준 모델의 MAE 계산하기**

```python
def evaluate_naive_method(dataset):
    total_abs_err = 0.
    samples_seen = 0
    for samples, targets in dataset:
        ## 온도 특성은 칼럼 인덱스 1에 있어, samples[:, -1, 1]이 입력 시퀀스에 있는 마지막 온도 측정값
        ## 특성을 정규화했기 때문에 온도를 섭시로 바구려면 표준 편차를 곱하고 평균을 더해야함
        preds = samples[:, -1, 1] * std[1] + mean[1]
        total_abs_err += np.sum(np.abs(preds - targets))
        samples_seen += samples.shape[0] ## sample.shape(256, 120, 14)
    return total_abs_err / samples_seen

print(f"검증 MAE: {evaluate_naive_method(val_dataset):.2f}") ## 검증 MAE: 2.44
print(f"테스트 MAE: {evaluate_naive_method(test_dataset):.2f}") ## 테스트 MAE: 2.62
```

### 기본적인 머신 러닝 모델 시도해 보기

**밀집 연결 모델 훈련하고 평가하기**

```python
from tensorflow import keras
from tensorflow.keras import layers

## 120 * 14개로 변환 -> 시간적 구조가 훼손
inputs = keras.Input(shape=(sequence_length, raw_data.shape[-1]))
x = layers.Flatten()(inputs)
x = layers.Dense(16, activation="relu")(x)
outputs = layers.Dense(1)(x)
model = keras.Model(inputs, outputs)

## 최상의 모델 저장
callbacks = [
    keras.callbacks.ModelCheckpoint("jena_dense.h5",
                                    save_best_only=True)
]
model.compile(optimizer="rmsprop", loss="mse", metrics=["mae"])
## timeseries_dataset_from_array()에 의해 (samples, targets) 형태의 튜플을 반복할 수 있는 데이터셋
history = model.fit(train_dataset,
                    epochs=10,
                    validation_data=val_dataset,
                    callbacks=callbacks)

## 최상의 모델을 다시 로드하고 테스트 데이터에서 평가
model = keras.models.load_model("jena_dense.h5")
print(f"테스트 MAE: {model.evaluate(test_dataset)[1]:.2f}")
```

**결과 그래프 그리기**

```python
import matplotlib.pyplot as plt
loss = history.history["mae"]
val_loss = history.history["val_mae"]
epochs = range(1, len(loss) + 1)
plt.figure()
plt.plot(epochs, loss, "bo", label="Training MAE")
plt.plot(epochs, val_loss, "b", label="Validation MAE")
plt.title("Training and validation MAE")
plt.legend()
plt.show()
```

## 순환 신경망 이해하기

**넘파이로 구현한 간단한 RNN**

```python
import numpy as np

timesteps = 100 ## 입력 시퀀스에 있는 타임스텝의 수
input_features = 32 ## 입력 특성의 차원
output_features = 64 ## 출력 특성의 차원
inputs = np.random.random((timesteps, input_features)) ## 입력 데이터: 에제를 위한 랜덤한 잡음
state_t = np.zeros((output_features,)) ## 초기 상태: 0백터
## 랜덤한 가중츠 행렬을 만듦
W = np.random.random((output_features, input_features))
U = np.random.random((output_features, output_features))
b = np.random.random((output_features,))
successive_outputs = []
## input_t은 (input_features,) 크기의 백터
for input_t in inputs:
    ## 입력과 현재 상태(이전 출력)를 연결하여 현재 출력을 얻음
    ## tanh 함수를 사용해 비선형성을 추가(다른 활성화 함수도 사용할 수 있음)
    output_t = np.tanh(np.dot(W, input_t) + np.dot(U, state_t) + b)
    successive_outputs.append(output_t) ## 출력을 리스트에 저장
    state_t = output_t ## 다음 타임스텝을 위해 네트워크의 상태를 업데이트
## 최종 출력: (timesteps, out_features)크기의 랭크-2 텐서
final_output_sequence = np.stack(successive_outputs, axis=0)
```
