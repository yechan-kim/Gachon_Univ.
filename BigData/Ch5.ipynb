# 조건을 이용하여 데이터 선택

```python
# 판다스 라이브러리 불러오기
import pandas as pd

# 시본 라이브러리 불러오기
import seaborn as sns

# tips 데이터셋 불러오기
tips = sns.load_dataset('tips')

# 전체 데이터 중 팁이 5달러 이상
tips[tips.tip > 5]
```

```python
# 남자 손님이면서 비흡연자
tips[(tips['sex']=='Male') & (tips['smoker']=='No')]
```

```python
# 남자 손님이거나 비흡연자
tips[(tips['sex']=='Male') | (tips['smoker']=='No')]
```

```python
# 여자 손님이면서 비흡연자
tips[~(tips['sex']=='Male') & (tips['smoker']=='No')]
```

```python
# 날짜가 일요일이면 True
[tips['day'].isin(['Sun'])]
```

```python
# 팁을 1달러 지불한 고객
tips[tips['tip'].isin([1])]
```

# 데이터 조작

```python
# 판다스 라이브러리 불러오기
import pandas as pd

# 시본 라이브러리 불러오기
import seaborn as sns

# tips 데이터셋 불러오기
tips = sns.load_dataset('tips')

#데이터 개수 확인
tips.count()
len(tips)
```

```python
# 인덱스 보기
tips.index
```

```python
# 열 보기
tips.columns
```

```python
# 행, 열 구조 보기
tips.values
```

```python
# 지급액 열(total_bill)을 기준으로 index(axis=0) 오름차순 정렬
# 행: aixs = 0
# 열: aixs = 1
tips.sort_values(by=['total_bill'], axis=0)
```

```python
# 지급액 열(total_bill)을 기준으로 index 내림차순 정렬
tips.sort_values(by=['total_bill'], axis=0, ascending=False)
```

```python
# 열 이름을 알파벳 순서로 정렬하기
tips.sort_index(axis=1)
```

```python
# 행 합계
# 행 합계를 계산할 때는 숫자형 데이터만을 고려하므로, 숫자가 아닌 데이터가 있는 열이 포함되어 있으면 오류 발생
tips.select_dtypes(include='number').sum(axis=1)
```

```python
# 열 합계
# 'Categorical' 데이터 타입은 일반적으로 범주형 데이터를 나타내는 데 사용되며, 합산 연산을 지원하지 않음
tips.select_dtypes(include='number').sum(axis=0)
tips.select_dtypes(include='number').sum()
```
