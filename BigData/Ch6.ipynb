```python
# 연습문제 6번

# 판다스 라이브러리 불러오기
import pandas as pd

# 맷플롯립 라이브러리 불러오기
import matplotlib as mpl
import matplotlib.pylab as plt

# 파일 경로를 찾고 변수 file_path에 저장
file_path = 'bicycle_out.csv'

# read csv() 함수로 데이터 프레임 변환
df = pd.read_csv(file_path, engine='python', encoding='euc-kr')

# 고급 시각화 박스플롯 참조
# 나이 이상치 시각화
plt.boxplot(df['나이'])
plt.show()

# 이상치 제거
Q1 = df['나이'].quantile(0.25)
Q3 = df['나이'].quantile(0.75)
IQR = Q3 - Q1

# 수염설정 (1.5 * IQR을 기준으로)
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 이상치 제거
clean_df = df[(df['나이'] > lower_bound) & (df['나이'] < upper_bound)]

# 이상치를 제거한 후에 재확인 
plt.boxplot(clean_df['나이'])
plt.show()
```

```python
# 판다스 라이브러리 불러오기
import pandas as pd

# 파일 경로를 찾고 변수 file_path에 저장
file_path = 'bicycle_dup.csv'

# read csv() 함수로 데이터 프레임 변환
df = pd.read_csv(file_path, engine='python', encoding='euc-kr')

# duplicated() 함수로 중복 데이터 찾기
# DataFrame.duplicated() 함수는 boolean 형태의 Series 반환
df.duplicated(['이용거리'])
```

```python
df.duplicated(['이용거리','나이'])
```

```python
# 중복이 있으면 처음이나 끝에 무엇을 남길지 확인함
# keep='first'가 default 값
# 중복같이 있으면 첫 번째 값을 duplicate 여부를 False로 반환, 나머지는 True로 반환
df.duplicated(['이용거리'], keep='first')
```

```python
# keep='last'를 사용하면 중복이 있으면 마지막 값을 duplicate 여부를 False로 반환, 나머지는 True로 반환
df.duplicated(['이용거리'], keep='last')
```

```python
# keep=False를 사용하면 중복이 있으면 모두 duplicate 여부를 True 반환
df.duplicated(['이용거리'], keep=False)
```

```python
# drop_duplicates() 함수는 중복값을 keep 옵션에 따라 유일한 1개의 키 값만 남기고 나머지는 중복 제거
df.drop_duplicates(['이용거리'], keep='first')
```

```python
df.drop_duplicates(['이용거리'], keep='last')
```

```python
df.drop_duplicates(['이용거리'], keep=False)
```
