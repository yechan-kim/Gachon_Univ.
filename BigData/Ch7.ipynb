```python
# 판다스 라이브러리 불러오기
import pandas as pd

# 파일 경로를 찾고 변수 file_path에 저장
file_path = 'exam_sample.csv'

# read csv() 함수로 데이터 프레임 변환
df = pd.read_csv(file_path, engine='python', encoding='euc-kr')

# math 변수에 대해서 3개 동일한 길이로 범주형 변수 생성
# 범주 구간은 아래와 같고, 3개 구간 길이가 동일함
# Categories (3, interval[float64]): [(19.93, 43.333] < (43.333, 66.667] < (66.667, 90.0]]
df1 = pd.cut(df.math, 3)
df1
```

```python
# math 변수에 3개 구간의 범주를 그룹 함수를 추가해서
# 각 범주의 그룹별로 agg() 함수인 개수(count), 평균(mean), 표준편차(std), 최솟값(min), 최댁값(max)를 계산
df2 = df.math.groupby(df1)
df3 = df2.agg(['count','mean','std','min','max'])
df3
```

```python
#pd.qcut() 함수를 사용해 math에 대해소 각 범주 구간별로 동일하게 3개의 개수를 가지도록 범주를 만들 수도 있음
# 이때 labels = False로 설정하여 label이 0, 1, 2 구간을 0부터 순차적으로 1씩 증가하게 합니다.
df4 = pd.qcut(df.math, 3, labels=False)
df4
```

```python
# 넘파이 라이브러리 불러오기
import numpy as np

# 아래처럼 labels = np.arange(3, 0, -1)로 직접 지정하면 label이 3, 2, 1로 3부터 1씩 줄어드는 순서로 할당이 됨
# 위의 label과 정반대로 할당됨
df5 = pd.qcut(df.math, 3, labels=np.arange(3, 0, -1))
df5
```

```python
## [3 < 2 < 1] 순서로 동일 개수로 나눈 3개의 그룹별 통계량을 걔산할 수 있음
df6 = df.math.groupby(df5)
df7 = df6.agg(['count','mean','std','min','max'])
df7
```
